```python
import requests

keyword = input("چه چیزی را می‌خواهی جستجو کنی؟ ")
pages = int(input("چند صفحه جستجو شود؟ "))

for page in range(1, pages + 1):
    url = f"https://api.digikala.com/v1/search/?q={keyword}&page={page}"


    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        products = data.get("data", {}).get("products", [])
        print(f"\n--- صفحه {page} ---")

        for i, product in enumerate(products, 1):
            title = product.get("title_fa", "بدون عنوان")
            price_data = product.get("default_variant", {}).get("price", {})
            price = price_data.get("selling_price", "نامشخص")

            if isinstance(price, int):
                price = f"{price:,} تومان"
            else:
                price = "نامشخص"

            print(f"{i}. {title} - {price}")
    else:
        print(f"خطا در دریافت اطلاعات از صفحه {page}")
```
