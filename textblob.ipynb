```python
from textblob import TextBlob
from deep_translator import GoogleTranslator

text = input("text?")
text = GoogleTranslator(source='auto', target='en').translate(text=text)
emotion = TextBlob(text)
sentiment = emotion.sentiment.polarity

print(text)
print(sentiment)

#sentiment > 1 : +
#sentiment == 0 : ~
#sentiment < 0 : -

if sentiment > 0 :
    print("+")
elif sentiment < 0 :
    print("-")
else :
    print("~")


```
