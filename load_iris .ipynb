```python
from sklearn.datasets import load_iris 
import pandas
import matplotlib.pyplot as plt

#load_iris : spesific dataset from the sklearn library about plants
#pandas : used for creating data frame
data = load_iris()
#"data" variable is the iris dataset
df = pandas.DataFrame(data.data, columns = data.feature_names )
#new variable named "df", filled with a data frame (created using "pandas"), gets data from the variable "data" (aka load_iris), and creates columns using the feature names from the "data" variable
df['target'] = data.target
#created new column named "target", which is filled with "target" from "data" variable (aka load_iris)
df['targate_name'] = [data.target_names[i] for i in data.target]
#created new column named "targate_name", which translates "data"s int data, to plant name(in the form of str)
df.head()
#"head" used to print data frame
from sklearn.model_selection import train_test_split
#train_test_split : splits 
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors = 3)
model.fit(X_train, y_train)

accuracy = model.score(X_test, y_test)
accuracy
```

```python
from sklearn.datasets import load_iris 
import pandas
import matplotlib.pyplot as plt

data = load_iris()
df = pandas.DataFrame(data.data, columns = data.feature_names )
df['target'] = data.target
df['targate_name'] = [data.target_names[i] for i in data.target]
df.head()

from sklearn.model_selection import train_test_split

X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors = 3)
model.fit(X_train, y_train)

accuracy = model.score(X_test, y_test)
accuracy
```
