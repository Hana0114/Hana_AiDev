```python
import cv2
import matplotlib.pyplot as plt
def ishow(img_webcam):
    cv2.imshow('img_1',img_webcam)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
    
```

```python
cam = cv2.VideoCapture(0)
_, img_webcam = cam.read()


img_webcam = cv2.cvtColor(img_webcam, cv2.COLOR_RGB2BGR)
ishow(img_webcam)
```
