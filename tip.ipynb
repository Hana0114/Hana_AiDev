```python
import seaborn as sb
from sklearn.preprocessing import LabelEncoder 
```

```python
df = sb.load_dataset("tips")
df
```

```python
encoder = LabelEncoder()
df['sex'] = encoder.fit_transform(df['sex'])

df['smoker'] = encoder.fit_transform(df['smoker'])

df['day'] = encoder.fit_transform(df['day'])

df['time'] = encoder.fit_transform(df['time'])

#use encoder to turn sring based data into intiger based data
df
```

```python
from sklearn.model_selection import train_test_split 
X = df.drop(['tip'], axis = 1)
#X is df dataframe but with dropped tip column. (column being shown by "axis = 1")
y = df['tip']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) 
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

model.score(X_test, y_test)
```

```python
total_bill_user = float(input("total bill?"))
sex_user = int(input("server sex?   (0 for female, 1 for male)"))
smoker_user = float(input("smoker?   (0 for no, 1 for yes)"))
day_user = float(input("day of the week?   (0 for sunday, 1 for monday,...)"))
time_user = float(input("time?   (0 for dinner, 1 for lunch,...)"))
size_user = float(input("size?"))

a = [[total_bill_user, sex_user, smoker_user, day_user, time_user, size_user]]
y_pred = model.predict(a)
print(y_pred)
```
