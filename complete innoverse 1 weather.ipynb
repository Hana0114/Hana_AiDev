```python
import numpy as np
import pandas as pd

np.random.seed(42)
#num_samples = 1000
num_samples = int(input())

def drought_risk(row):
    if row['precipitation'] < 400 and row['water_level'] < 30 and row['temperature'] > 30:
        return 2
    elif row['precipitation'] < 500 or row['water_level'] < 50:
        return 1
    else:
        return 0


data = pd.DataFrame({
    'region': [f'Region-{i}' for i in range(num_samples)],
    'temperature': np.random.normal(loc = 30, scale = 5, size = num_samples),
    'precipitation': np.random.normal(loc = 500, scale = 100, size = num_samples),
    'humidity': np.random.normal(loc = 50, scale = 10, size = num_samples),
    'water_level': np.random.uniform(0, 100, size = num_samples),
    'cultivated_area': np.random.uniform(0.1, 0.9, size = num_samples)
})


data['risk'] = data.apply(drought_risk, axis=1)

data.head

from sklearn.ensemble import RandomForestClassifier as RandomForest
from sklearn.model_selection import train_test_split 

feature = ['temperature', 'precipitation', 'water_level', 'humidity', 'cultivated_area'] 
X = data[feature]
y = data['risk']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.2, stratify= y)
model = RandomForest(n_estimators= 100)
```
